```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.model_selection as train_test_split
import sklearn.linear_model as linear_model
```

```python
# Load the data
data = pd.read_csv('SeoulBikeDataClean.csv')
```

## Modelo de regresión

```python
indice = 0
for i in data.columns:
    if i != 'Rented Bike Count':
        print(f"X{indice}:", i)
    else:
        print("Y: ",i)
    indice += 1
    
```

```python
# Separar X y Y
X = data.drop('Rented Bike Count', axis=1)
Y = data['Rented Bike Count']
X.head()
```

```python
Y.head()
```

### División entre entrenamiento y prueba

```python
import sklearn.model_selection as train_test_split

# Dividir los datos en entrenamiento y prueba
X_train, X_test, Y_train, Y_test = train_test_split.train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
# Tamaño de los datos de entrenamiento y prueba
print("Tamaño de los datos de entrenamiento X:", X_train.shape)
print("Tamaño de los datos de prueba X:", X_test.shape)
print("Tamaño de los datos de entrenamiento Y:", Y_train.shape)
print("Tamaño de los datos de prueba Y:", Y_test.shape)
```

### Modelo RL

```python
# Crear el modelo de regresión lineal
import sklearn.linear_model as lm
model = lm.LinearRegression()
model.fit(X_train, Y_train)
```

```python
# Imprimir coeficientes
variables = X.columns
print("B0:", model.intercept_)
indice = 1
for i in model.coef_:
    feature = variables[indice-1]  
    print(f"B{indice} ({feature}):", i)
    indice += 1
```

### Predicciones datos de prueba

```python
y_pred = model.predict(X_test)
```

### Evaluar el modelo

**Error absoluto medio**: 

$$\text{MAE} = \frac 1n\sum_{i=1}^n|y_i-\hat{y}_i|$$

**Error cuadrado medio**: 
$$\text{MSE} = \frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2$$

**Raíz del Error cuadrado medio**: 
$$\text{RMSE} = \sqrt{\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2}$$

```python
# Evaluar el modelo
import numpy as np
import sklearn.metrics as metrics

# Calcular el absoluto medio
print("MAE:", metrics.mean_absolute_error(Y_test, y_pred))

# Calcular el error cuadrático medio
print("MSE:", metrics.mean_squared_error(Y_test, y_pred))

# Raíz del error cuadrático medio
print("RMSE:", np.sqrt(metrics.mean_squared_error(Y_test, y_pred)))

# R cuadrado
print("R2:", metrics.r2_score(Y_test, y_pred))
```

### Validacion Cruzada

```python
# Validación cruzada
import sklearn.model_selection as model_selection
MAE_scores = model_selection.cross_val_score(model, X, Y, cv=5, scoring='neg_mean_absolute_error')
MSE_scores = model_selection.cross_val_score(model, X, Y, cv=5, scoring='neg_mean_squared_error')
RMSE_scores = model_selection.cross_val_score(model, X, Y, cv=5, scoring='neg_root_mean_squared_error')
R2_scores = model_selection.cross_val_score(model, X, Y, cv=5, scoring='r2')

# Imprimir los resultados
print("MAE:", -MAE_scores)
print("MAE promedio:", np.mean(-MAE_scores))

print("\nMSE:", -MSE_scores)
print("MSE promedio:", np.mean(-MSE_scores))

print("\nRMSE:", -RMSE_scores)
print("RMSE promedio:", np.mean(-RMSE_scores))

print("\nR2:", R2_scores)
print("R2 promedio:", np.mean(R2_scores))

```

### Sugerencias para Mejorar el Modelo

Verificar y Manejar la Multicolinealidad:

Utiliza el VIF (Variance Inflation Factor) para identificar y tratar la multicolinealidad. Si alguna de las variables tiene un VIF alto (mayor a 10), podrías considerar eliminarla o combinarla con otras variables correlacionadas.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Calcular VIF para cada variable independiente
X = data[['Hour', 'Temperature(C)', 'Humidity(%)', 'Wind speed (m/s)', 'Dew point temperature(C)',
          'Solar Radiation (MJ/m2)', 'Rainfall(mm)', 'Snowfall (cm)', 'Holiday', 'Functioning Day',
          'Seasons_Spring', 'Seasons_Summer', 'Seasons_Winter']]

vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]

print(vif_data)
```
